```python
import pandas as pd
from binance.client import Client
```

```python
# insert API key / API Secret

api_key = ""
api_secret = ""
```

```python
# connect to binance client
client = Client(api_key, api_secret)
```

```python
# get candle from 1 Nov 2017 to 1 Nov 2021
BTCcandle = client.get_historical_klines("MANAUSDT", Client.KLINE_INTERVAL_1DAY, "1 Nov, 2017", "1 Nov, 2021")
```

```python
# import into pandas dataframe
btc_df = pd.DataFrame(BTCcandle)
btc_df.columns = ['Open Time', 'Open', 'High', 'Low', 'Close', 'Volume', 'Close Time', 'Quote Asset Volume', 'No. Trade', 'TB Base Volume', 'TB Quote Volume', 'Ignore']
```

```python
btc_df
```

```python
btc_df = btc_df[['Close Time', 'Close', 'Volume', 'Quote Asset Volume']]

btc_df['Close Time'] = pd.to_datetime(btc_df['Close Time']/1000, unit = 's')
```

```python
btc_df['Close Time'] = btc_df['Close Time'].dt.date
```

```python
btc_df['Close'] = btc_df['Close'].astype(float)
btc_df['Close'].round(2)

btc_df
```

```python
btc_df.to_csv("SAND_NOV2017_NOV2021.csv")
```
